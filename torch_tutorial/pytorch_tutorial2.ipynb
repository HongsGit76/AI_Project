```python
# 만약 .requires_grad 속성을 True 로 설정하면, 
# 그 tensor에서 이뤄진 모든 연산들을 추적(track)하기 시작합니다.
import torch

# 연산을 기록
x = torch.ones(2,2, requires_grad=True)
print(x)
```

```python
y = x+2
print(y)
```

```python
# y는 연산의 결과로 생성된 것이므로 grad_fn을 갖는다.
print(y.grad_fn)
```

```python
z = y * y *3
out = z.mean()

print(z, out)
```

```python
a = torch.randn(2,2)
a = ((a * 3) / (a - 1))
print(a.requires_grad)
a.requires_grad_(True)
print(a.requires_grad)
b = (a * a).sum()
print(b.grad_fn)
```

```python
out.backward()
```

```python
# 변화도 d(out)/dx를 출력
print(x.grad)
```

```python
# 벡터 야코비안 곱의 예제를 살펴보도록 하겠습니다.
x = torch.randn(3, requires_grad=True)

y = x * 2
while y.data.norm() < 1000:
  y = y*2

print(y)
```

```python
v = torch.tensor([0.1, 1.0, 0.0001], dtype=torch.float)
y.backward(v)

print(x.grad)
```

```python
print(x.requires_grad)
print((x ** 2).requires_grad)

with torch.no_grad():
  print((x**2).requires_grad)
```

```python
print(x.requires_grad)
y = x.detach()
print(y.requires_grad)
print(x.eq(y).all())
```
