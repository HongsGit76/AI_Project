```python
from __future__ import print_function
import torch
```

```python
# 초기화 되지 않은 행렬 생성
x = torch.empty(5,3)
print(x)
```

```python
# 무작위로 초기화 한 행렬 생성
x = torch.rand(5,3)
print(x)
```

```python
#dtype이 long이고, 0으로 채워진 행렬 생성 
x = torch.zeros(5,3,dtype=torch.long)
print(x)
```

```python
# 데이터로부터 tensor직접 생성
x = torch.tensor([5.5, 3])
print(x)
```

```python
# 기존 tensor를 바탕으로 새로운 tensor 생성
x = x.new_ones(5,3,dtype=torch.double)
print(x)

x = torch.randn_like(x, dtype=torch.float)
print(x)
```

```python
# 행렬의 크기 구하기
print(x.size())
```

```python
# 행렬 덧셈 1
y = torch.rand(5,3)
print(x + y)
```

```python
# 행렬 덧셈 2
print(torch.add(x,y))
```

```python
# 행렬 덧셈 3
result = torch.empty(5,3)
torch.add(x, y, out=result)
print(result)
```

```python
# y에 x 더하기
y.add_(x)
print(y)
```

```python
print(x[:,1])
```

```python
# 텐서의 크기, 모양 변경할 때 torch.view 사용
x = torch.randn(4,4)
y = x.view(16)
z = x.view(-1,8) # -1은 다른 차원에서 유추
print(x.size(), y.size(), z.size())
```

```python
# tensor에 하나의 값만 존재한다면 .item()으로 숫자값얻기 가능
x = torch.randn(1)
print(x)
print(x.item())
```

```python
# tensor(torch) -> array(numpy)
a = torch.ones(5)
print(a)

b = a.numpy()
print(b)
```

```python
a.add_(1)
print(a)
print(b)
```

```python
# tensor(torch) <- array(numpy)
import numpy as np
a = np.ones(5)
b = torch.from_numpy(a)
np.add(a,1,out=a)
print(a)
print(b)
```
